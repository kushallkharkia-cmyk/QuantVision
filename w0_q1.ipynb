```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
import matplotlib.pyplot as plt
```

```python
# Exercise 1
fig= plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
```

```python
# Exercise 2
fig2 = plt.figure()

ax1 = fig2.add_axes([0, 0, 1, 1])
ax2 = fig2.add_axes([0.2, 0.5, 0.2, 0.2])
```

```python
fig2 = plt.figure()

ax1 = fig2.add_axes([0, 0, 1, 1])
ax2 = fig2.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y)
ax1.set_title("ax1")
ax2.plot(x, y)
ax2.set_title("ax2")
plt.show()
```

```python
# Exercise 3
fig3=plt.figure()
a1=fig3.add_axes([0,0,1,1])
a2=fig3.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig3=plt.figure()
a1=fig3.add_axes([0,0,1,1])
a2=fig3.add_axes([0.2,0.5,0.4,0.4])
a1.plot(x,z)
a2.plot(x,y)
a2.set_title("zoom")
a1.set_xlabel("X")
a1.set_ylabel("Z")
a2.set_xlabel("X")
a2.set_ylabel("Y")
a2.set_xlim(20,22)
a2.set_ylim(30,50)
a1.set_ylim(0,10000)
a1.set_xlim(0,100)
plt.show()
```
