```python
import yfinance as yf
ticker = 'GS'
stock = yf.Ticker(ticker)
df = stock.history(period='6mo', interval='1d')
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Candlestick(
    x=df.index,
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close'],
    name=ticker
)])
fig.update_layout(
    title=f'Goldman Sachs ({ticker}) - 6 Month History',
    yaxis_title='Stock Price (USD)',
    xaxis_title='Date',
    xaxis_rangeslider_visible=False,
)
fig.show()
```

1. 25-26 August,2025
2. Bullish Engulf
3. The closing price of next day(27th August, 2025) increased than the closing of 26th.
